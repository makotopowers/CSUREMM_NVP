# Sparsity & Costs

Idea: many SKUs are ordered only a handful of times throughout the month. Most days, that SKU has an order quantity of zero. We call these types of SKUs sparse. It is very difficult to predict the order quantity for sparse SKUs and we try to quantify the following:
1. How many SKUs are sparse in the JD dataset?
2. How difficult is it to predict the order quantity for sparse SKUs?
3. Are many of the costs we incurr due to mistakes made on sparse SKUs?

```python
# For each SKU, compute the percent of orders that are 0 for each SKU
sparsity = (orders == 0).sum(axis=1) / orders.shape[1]
print('The average SKU has {:.3}% of days with 0 orders placed (std = {:.3}).'.format(sparsity.mean() * 100, sparsity.std() * 100))

fig, ax = plt.subplots(figsize=(10, 5))
ax.hist(sparsity, bins=10, label='Percent sparsity')
ax.set(xlabel='Percentage of days that are sparse for a given SKU', ylabel='Count', title='SKU sparsity distribution')
ax.legend()

plt.show()
```

Let's take a closer look at indivudal SKUs to get a sense of how sparse they are:

```python
def plot_sku(ts, sku_id, sparse=None, varience=None):
    fig, ax = plt.subplots(figsize=(5, 3))
    ax.bar(range(len(ts)), ts)
    sparse_msg = '({:.3}% Sparse)'.format(sparse * 100) if sparse is not None else ''
    varience_msg = '(Var = {:.3})'.format(varience) if varience is not None else ''
    ax.set(xlabel='Date', ylabel='Quantity', title='SKU {}'.format(sku_id) + sparse_msg + varience_msg)
    plt.show()
```

### Cost Analysis

```python
def vectorized_SAA(orders, beta):
    n = orders.shape[1]
    df = orders.apply(lambda x: np.sort(x), axis=1, raw=True)
    idx = int(np.ceil(n * beta))-1
    y_pred = df.iloc[:, idx].to_numpy()
    return y_pred
```

```python
def get_cost(y_pred, y_true, c_o, c_u):
    """ y_pred = predicted demand for each day; y_true = actual demand for each day """
    return c_u * np.squeeze((y_true-y_pred)*[y_true > y_pred]) + c_o * np.squeeze((y_pred-y_true) * [y_pred > y_true])
```

```python
c_u, c_o = 2, 2
beta = c_u / (c_u + c_o)

predictions = np.zeros(orders.shape)
costs = np.zeros(orders.shape)

for day_idx in range(1, orders.shape[1]):

    y_pred = vectorized_SAA(orders.iloc[:, :day_idx], beta)
    y_true = orders.iloc[:, day_idx].to_numpy()
    cost = get_cost(y_pred, y_true, c_o, c_u)
    
    predictions[:, day_idx] = y_pred
    costs[:, day_idx] = cost
```

```python
def get_avg_cost(orders, c_u, c_o):
    beta = c_u / (c_u + c_o)
    
    costs = []
    
    for day_idx in range(1, orders.shape[1]):

        y_pred = vectorized_SAA(orders.iloc[:, :day_idx], beta)
        y_true = orders.iloc[:, day_idx].to_numpy()
        cost = get_cost(y_pred, y_true, c_o, c_u)
        costs.append(cost)
        
    avg_cost = np.mean(costs, axis=0)
    return avg_cost
```

```python
avg_cost = get_avg_cost(orders, c_u=1, c_o=1)
orders['avg_cost'] = avg_cost
```

# Newsvendor Analysis

An analysis of the newsvendor problem.

In our analysis of the newsvendor problem, there are several scenarios
that we want to consider when picking the optimal algorithm.

1.  Is the data stationary or non-stationary?
2.  If non-stationary, what kind of non-stationary?
3.  For each environment (stationary, non-stationary 1, non-stationary
    2), what is the optimal algorithm?

We first look at the stationary environment. If the data is stationary,
then by the law of large numbers we want to consider as many data points
as possible. We run our ranking algorithm and get the following result.
We choose our data to be close to stationary by running preliminary
preprocessing techniques, like removing outlier points.

For these algorithms we assumed that the distribution is stationary. We
would also like to have a way to determine if the data is stationary or
non-stationary. We do this using change-point detection. Mathematically,
we can describe a time series as an ordered sequence of observations
$(x_1, x_2, \ldots )$. We can write the joint distribution of a
subset $x_{a:b} = (x_a, x_{a+1}, \ldots,x_b)$ of the time series as
$p(x_{a:b})$. If the goal is to determine whether a change point
occurred at a time $\tau$ in a finite time series of length $T$,
then we really ask whether $p(x_{1:\\tau })$ equals $p(x_{\tau
+1:T})$. This problem can be generalized to the case of more than one
change point.

If we decide that the data is non-stationary, how can we classify the
shift that we see? Is it a slow, constant drift, or arbitrary change?

We can see this doesn't work. The overlap indicates obscure trends of each cluster.

We decompose each time series into three parts: trend, seasonal, and residuals. Instead of using the full time series, we are using the variances of the three decomposed time series as features of clustering.

```python
def decompose(observed):
    #result = seasonal_decompose(observed, model='additive', period = 7, extrapolate_trend = 'freq', filt = 0.5 + np.zeros(len(observed)))
    result = STL(observed, period = 7, robust = True, seasonal_jump = 150, trend_jump = 150).fit()
    return np.array([stat.variance(result.trend), stat.variance(result.seasonal), stat.variance(result.resid)])/stat.variance(result.observed)
```

```python
decomposed_vectors = np.array(list(map(decompose, vectors)))

clustering = SpectralClustering(n_clusters= 2,assign_labels='discretize', random_state=0).fit(decomposed_vectors)
plt.plot(np.add.reduce(vectors[np.where(clustering.labels_ == 0)[0]])/len(np.where(clustering.labels_ == 0)[0]), label = '0')
plt.plot(np.add.reduce(vectors[np.where(clustering.labels_ == 1)[0]])/len(np.where(clustering.labels_ == 1)[0]), label = '1')

plt.legend()
plt.show()
```

```python
cluster_no = 3
clustering = SpectralClustering(n_clusters= cluster_no,assign_labels='discretize', random_state=0).fit(decomposed_vectors)

for i in range(cluster_no):
    plt.plot(np.add.reduce(vectors[np.where(clustering.labels_ == i)[0]])/len(np.where(clustering.labels_ == i)[0]), label = i)

plt.legend()
plt.show()
```
